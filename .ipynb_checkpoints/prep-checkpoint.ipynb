```python
state = gpd.read_file(r"data\raw\cb_2018_us_state_500k\cb_2018_us_state_500k.shp")
county = gpd.read_file(r"data\raw\nhgis0004_shapefile_tl2021_us_county_2021\US_county_2021.shp")
stations = pd.read_csv(r"data\created\stations_20250130.csv")
stations = gpd.GeoDataFrame(stations, geometry=gpd.points_from_xy(stations['longitude'], stations['latitude']), crs="EPSG:4269")
county = county.to_crs(epsg=4269)
```

```python
# Define the bounding box for the lower 48 states
min_longitude = -125.0
max_longitude = -66.93457
min_latitude = 24.396308
max_latitude = 49.384358

# Filter the state GeoDataFrame for the lower 48 states using the bounding box
lower_48_states = state.cx[min_longitude:max_longitude, min_latitude:max_latitude]
lower_48_counties = county.cx[min_longitude:max_longitude, min_latitude:max_latitude]

# Filter the stations GeoDataFrame for stations within the bounding box
stations_filtered = stations.cx[min_longitude:max_longitude, min_latitude:max_latitude]

# Keep either DCFC or L2 stations, remove free stations, or high outliers
stations_filtered = stations_filtered[(stations_filtered['dcfc'] == 0)&(stations_filtered['station_rate'] > 0)&(stations_filtered['station_rate'] < stations_filtered['station_rate'].quantile(0.99))]

bins = [0.00, 0.10, 0.20, 0.30, 0.40, 0.50, float('inf')]
labels = ['0.00-0.09', '0.10-0.19', '0.20-0.29', '0.30-0.39', '0.40-0.49', '0.50+']
```

```python
lower_48_states["geometry"] = lower_48_states["geometry"].simplify(tolerance=0.01, preserve_topology=True)
stations_filtered["geometry"] = stations_filtered["geometry"].simplify(tolerance=0.001)
county["geometry"] = county["geometry"].simplify(tolerance=0.001)

stations_filtered['rate_binned'] = pd.cut(stations_filtered['station_rate'], bins=bins, labels=labels, right=False)

# Count the number of stations in each bin
bin_counts = stations_filtered['rate_binned'].value_counts().sort_index()

# Create new labels with the count of stations in each bin, with custom space for 0.00 and 1.00+
max_length = max(len(label) for label in labels)  # Find the longest label to align the counts
legend_labels = []
for label in labels:
    count = bin_counts[label]
    if label == '0.50+':  # Add 5 spaces for this label
        legend_labels.append(f"{label}{' ' * (max_length - len(label) + 6)}({count})")
    else:  # Add 4 spaces for other labels
        legend_labels.append(f"{label}{' ' * (max_length - len(label) + 4)}({count})")

# Plot the lower 48 states with thinner boundaries
fig, ax = plt.subplots(figsize=(10, 10))
lower_48_states.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.2)  # Set thinner boundary lines

# Create a colormap for the binned categories
cmap = mpl.cm.YlOrRd  # You can choose other colormaps

# Plot the stations with the 0.00 bin first (lighter colors)
stations_filtered[stations_filtered['rate_binned'] != '0.00'].plot(ax=ax, marker='o', column='rate_binned', cmap=cmap, 
    markersize=8, legend=True, legend_kwds={'labels': legend_labels, 'bbox_to_anchor': (1, 0), 'loc': 'lower right', 'title': "Station\nPrice ($/kWh)"}, alpha=0.7)


# Remove grid lines
ax.axis(False)

plt.savefig("small_map_l2_price_stations.svg", format="svg")

# Show the plot
plt.show()
```

```python
stations_filtered['STATEFP'] = stations_filtered['fips_state'].astype(str).str.zfill(2)
stations_filtered['COUNTYFP'] = stations_filtered['fips_county'].astype(str).str.zfill(3)

# Merge the stations with the county DataFrame based on STATEFP and COUNTYFP
merged = stations_filtered.merge(county, how="left", on=["STATEFP", "COUNTYFP"])

# Calculate the average rate per county
county_avg_rate = merged.groupby('GEOID')['station_rate'].mean()

# Merge the average rate back to the county GeoDataFrame
county['avg_rate'] = county['GEOID'].map(county_avg_rate)

# Bin the 'avg_rate' into discrete ranges
county['avg_rate_binned'] = pd.cut(county['avg_rate'], bins=bins, labels=labels, right=False)

# Plot the counties with the binned average rates color-coded
fig, ax = plt.subplots(figsize=(10, 10))
lower_48_states.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.2)  # Set thinner boundary lines
county.plot(ax=ax, column='avg_rate_binned', cmap='YlOrRd', legend=True, 
            edgecolor='black', linewidth=0.2, facecolor='lightgrey', 
            legend_kwds={'bbox_to_anchor': (1, 0), 'loc': 'lower right', 'title': "Avg Station\nPrice ($/kWh)"})  # Position the legend

# Remove grid lines
ax.axis('off')

plt.savefig("small_map_l2_price_counties.svg", format="svg")

# Show the plot
plt.show()
```

```python
# Merge the stations with the county DataFrame based on STATEFP and COUNTYFP
merged = stations_filtered.merge(state, how="left", on=["STATEFP"])

# Calculate the average rate per county
state_avg_rate = merged.groupby('GEOID')['station_rate'].mean()

# Merge the average rate back to the state GeoDataFrame
state['avg_rate'] = state['GEOID'].map(state_avg_rate)

# Bin the 'avg_rate' into discrete ranges
state['avg_rate_binned'] = pd.cut(state['avg_rate'], bins=bins, labels=labels, right=False)

# Plot the states with the binned average rates color-coded
fig, ax = plt.subplots(figsize=(10, 10))
lower_48_states.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.2)  # Set thinner boundary lines
state.plot(ax=ax, column='avg_rate_binned', cmap='YlOrRd', legend=True, 
            edgecolor='black', linewidth=0.2, facecolor='lightgrey', 
            legend_kwds={'bbox_to_anchor': (1, 0), 'loc': 'lower right', 'title': "Avg Station\nPrice ($/kWh)"})  # Set legend position

# Remove grid lines
ax.axis('off')

plt.savefig("map_dcfc_price_states.svg", format="svg")


# Show the plot
plt.show()
```
